Import the Libraries

Import the main DataFrame

Show the head data

```python
import pandas as pd
aluguel_casa=pd.read_excel("houses_rent.xlsx")
aluguel_casa.head()
```

Show the tail data

```python
aluguel_casa.tail()
```

Use Rename to fix the lables

```python
aluguel_casa=aluguel_casa.rename(columns={"parking spaces":"parking_spaces", "rent amount":"rent_amount","property tax":"property_tax","fire insurance":"fire_insurance"})
aluguel_casa.head()
```

```python
aluguel_casa.dtypes
```

Replace the yes or no question by a bollean


```python
aluguel_casa['animal']=aluguel_casa['animal'].map({'not accept':False, 'accept': True})
#aluguel_casa['animal']=aluguel_casa['animal'].map('accept','True')
aluguel_casa['furniture']=aluguel_casa['furniture'].map({'not furnished':False,'furnished':True})
#aluguel_casa['furniture']=aluguel_casa['furniture'].str.replace('furnished','True')
aluguel_casa['floor'] = aluguel_casa['floor'].replace('-','0')
aluguel_casa.head()
```

Change data types

```python
aluguel_casa['floor'].info()
```

```python
aluguel_casa=aluguel_casa.astype({'animal':'bool','furniture':'bool'})
aluguel_casa=aluguel_casa.astype({'rent_amount':'float','property_tax':'float','fire_insurance':'float','property_tax':'float'})
aluguel_casa=aluguel_casa.astype({'floor':'int'})
aluguel_casa.dtypes
```

Aggregating data

the *groupby()* function groups the data in the DataFrame by the "city" column. This essentially creates separate groups of data for each unique city in the dataset.

*.agg("mean")* calculates the mean (average) of the "total" column for each city group. So, it's finding the average total cost for each city

*.reset_index()* resets the index of the resulting DataFrame. After the groupby() operation, the "city" becomes the index. *reset_index()* makes "city" a regular column again and creates a new numerical index.

```python
agg_city_price = aluguel_casa.groupby(["city"])['total'].agg("mean").reset_index()
agg_city_price
```

using groupby with 2 paramters

```python
agg_bath_room_price=aluguel_casa.groupby(["bathroom","rooms"])['total'].agg("mean").reset_index()
agg_bath_room_price
```
